```python
import pandas as pd
from pathlib import Path
import re
import numpy as np
```

```python
dataset_path = Path(r"../data/blue_waters_posix_with_paths_no_negative_outliers.csv")
```

```python
df = pd.read_csv(dataset_path)
```

```python
df.loc[df.path.isnull()]
```

```python
filename = Path(df.path[0]).parts[-1]
filename
```

```python
df
```

```python
re.search(r"_.*_id",filename).group()[1:-3]
```

Replace path column with app column

```python
df["app"] = df.path.astype("string").apply(lambda x: re.search(r"_.*id",Path(x).parts[-1]).group()[1:-3])
df = df.drop(["path"],axis=1)
df
```

```python
df.to_csv(r"../data/blue_waters_posix_with_apps_no_negative_outliers.csv",index=False)
```
